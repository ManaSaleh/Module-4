```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, 2:4]  # Selecting only petal length and petal width

# Define the lower and upper bounds for truncation
lower_bound = X.min(axis=0)  # Lower bound: minimum value for each feature
upper_bound = np.percentile(X, 95, axis=0)  # Upper bound: 95th percentile for each feature

# Truncate the outliers to the specified bounds
X_truncated = np.clip(X, lower_bound, upper_bound)

# Scale the truncated data using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_truncated)

# Initialize KMeans with the number of clusters (k) set to 3
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit KMeans to the scaled data
kmeans.fit(X_scaled)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Get the cluster labels for each data point
labels = kmeans.labels_

# Plotting the clusters
plt.figure(figsize=(8, 6))

# Scatter plot for petal length and petal width
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='red', s=200, label='Centroids')
plt.xlabel('Petal Length (scaled)')
plt.ylabel('Petal Width (scaled)')
plt.title('K-means Clustering of Scaled Iris Dataset (Petal Length vs Petal Width)')
plt.legend()
plt.show()
```
